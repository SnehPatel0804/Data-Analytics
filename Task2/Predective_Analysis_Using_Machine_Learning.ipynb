##Feature Selection


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Replace 'diagnosis' with your actual target column name
target_column = "diagnosis"

# Drop the target column from the dataset
X = df.drop(columns=[target_column, 'id'])  # Drop 'id' as it doesn't provide predictive value
y = df[target_column]

# Fit a Random Forest Classifier for feature importance
rf_model = RandomForestClassifier()
rf_model.fit(X, y)

# Get feature importances and plot top 10
feat_imp = pd.Series(rf_model.feature_importances_, index=X.columns)
feat_imp.nlargest(10).plot(kind='barh', title="Top 10 Feature Importances")
plt.show()
```

## Model Training :- A) Regression Models



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder

# Label encode the target variable 'diagnosis'
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # This will convert 'B' -> 0, 'M' -> 1

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
# Print the shapes of the train and test datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Train models
lr = LinearRegression().fit(X_train, y_train)
rf = RandomForestRegressor().fit(X_train, y_train)
xgb = XGBRegressor().fit(X_train, y_train)

# Print a confirmation message
print("Models trained successfully!")
```

```python
# Predictions
lr_pred = lr.predict(X_test)
rf_pred = rf.predict(X_test)
xgb_pred = xgb.predict(X_test)

# Print a confirmation message for predictions
print("Predictions made successfully!")

# Print some of the predictions to inspect them
print("Linear Regression Predictions (first 5):", lr_pred[:5])
print("Random Forest Predictions (first 5):", rf_pred[:5])
print("XGBoost Predictions (first 5):", xgb_pred[:5])
```

##Model Traning :- B) Classification Models (Using Binarized Target)

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Label encode the target column (if it's categorical)
label_encoder = LabelEncoder()
df["encoded_target"] = label_encoder.fit_transform(df[target_column])

# Binarize based on median
median_value = df["encoded_target"].median()
df["class_target"] = (df["encoded_target"] > median_value).astype(int)

# Prepare features and target for classification
X_cls = df.drop(columns=[target_column, "encoded_target", "class_target"])
y_cls = df["class_target"]

# Split classification dataset
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_cls, y_cls, test_size=0.2, random_state=42)

# Print the shapes of the train and test datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Train classification models
knn = KNeighborsClassifier().fit(X_train_cls, y_train_cls)
logreg = LogisticRegression(max_iter=1000).fit(X_train_cls, y_train_cls)
rfc = RandomForestClassifier().fit(X_train_cls, y_train_cls)
xgbc = XGBClassifier().fit(X_train_cls, y_train_cls)

# Print a confirmation message
print("Models trained successfully!")
```

```python
# Accuracy scores
knn_acc = accuracy_score(y_test_cls, knn.predict(X_test_cls))
logreg_acc = accuracy_score(y_test_cls, logreg.predict(X_test_cls))
rfc_acc = accuracy_score(y_test_cls, rfc.predict(X_test_cls))
xgbc_acc = accuracy_score(y_test_cls, xgbc.predict(X_test_cls))

# Print accuracy scores for each model
print(f"KNN Accuracy: {knn_acc}")
print(f"Logistic Regression Accuracy: {logreg_acc}")
print(f"Random Forest Accuracy: {rfc_acc}")
print(f"XGBoost Accuracy: {xgbc_acc}")
```

## Model Evaluation :- A) Regression Metrics


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluate_reg(name, y_true, y_pred):
    print(f"{name} R² Score: {r2_score(y_true, y_pred):.4f}")
    print(f"{name} MAE: {mean_absolute_error(y_true, y_pred):.4f}")
    print(f"{name} MSE: {mean_squared_error(y_true, y_pred):.4f}")
    print("-" * 40)

evaluate_reg("Linear Regression", y_test, lr_pred)
evaluate_reg("Random Forest Regressor", y_test, rf_pred)
evaluate_reg("XGBoost Regressor", y_test, xgb_pred)
```

## Model Evaluation :- B)  Classification Accuracy

```python
print(f"KNN Accuracy: {knn_acc:.4f}")
print(f"Logistic Regression Accuracy: {logreg_acc:.4f}")
print(f"Random Forest Classifier Accuracy: {rfc_acc:.4f}")
print(f"XGBoost Classifier Accuracy: {xgbc_acc:.4f}")
```

##Visualizations :- Classification

***A) Bar Chart for Classification Accuracies***

```python
models_cls = ['KNN', 'Logistic Regression', 'Random Forest', 'XGBoost']
accs_cls = [knn_acc, logreg_acc, rfc_acc, xgbc_acc]

plt.figure(figsize=(8, 5))
sns.barplot(x=models_cls, y=accs_cls)
plt.title("Classification Model Accuracies")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()
```

***B) Confusion Matrices for All Classifiers***

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

models = [("KNN", knn), ("Logistic Regression", logreg), ("Random Forest", rfc), ("XGBoost", xgbc)]

for name, model in models:
    plt.figure(figsize=(4, 4))
    cm = confusion_matrix(y_test_cls, model.predict(X_test_cls))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
    disp.plot(cmap="Blues", values_format='d')
    plt.title(f"{name} - Confusion Matrix")
    plt.show()
```

***C) ROC Curves for Classifiers***

```python
from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(8, 6))

for name, model in models:
    y_proba = model.predict_proba(X_test_cls)[:, 1]
    fpr, tpr, _ = roc_curve(y_test_cls, y_proba)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc(fpr, tpr):.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves for Classification Models")
plt.legend()
plt.grid()
plt.show()
```

## Visualization :- Regression


***A) Bar Chart for Regression R² Scores***

```python
r2_scores = [
    r2_score(y_test, lr_pred),
    r2_score(y_test, rf_pred),
    r2_score(y_test, xgb_pred)
]
models_reg = ['Linear Regression', 'Random Forest', 'XGBoost']

plt.figure(figsize=(8, 5))
sns.barplot(x=models_reg, y=r2_scores)
plt.title("Regression Model R² Scores")
plt.ylabel("R² Score")
plt.ylim(0, 1)
plt.show()
```

***B) Actual vs. Predicted Scatter Plots***

```python
reg_preds = [("Linear Regression", lr_pred), ("Random Forest", rf_pred), ("XGBoost", xgb_pred)]
for name, pred in reg_preds:
    plt.figure(figsize=(6, 5))
    sns.scatterplot(x=y_test, y=pred, alpha=0.6)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    plt.title(f"{name} - Actual vs Predicted")
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.grid(True)
    plt.show()
```

***C) Residual Plots***

```python
for name, pred in reg_preds:
    residuals = y_test - pred
    plt.figure(figsize=(6, 4))
    sns.histplot(residuals, kde=True, bins=30)
    plt.title(f"{name} - Residual Distribution")
    plt.xlabel("Residuals")
    plt.grid(True)
    plt.show()
```

***C) Feature Importances (for Tree Models)***

```python
def plot_feature_importance(model, model_name):
    importances = model.feature_importances_
    features = X_cls.columns
    sorted_idx = importances.argsort()[::-1]
    plt.figure(figsize=(8, 5))
    sns.barplot(x=importances[sorted_idx], y=features[sorted_idx])
    plt.title(f"{model_name} - Feature Importance")
    plt.xlabel("Importance")
    plt.show()

plot_feature_importance(rfc, "Random Forest")
plot_feature_importance(xgbc, "XGBoost (Classifier)")
```

## Overall Comparison of Classification and Regression Model Performances

```python
# Data
final_models = models_cls + models_reg + ['Avg Classification', 'Avg Regression']
final_scores = accs_cls + r2_scores + [sum(accs_cls)/len(accs_cls), sum(r2_scores)/len(r2_scores)]

# Assign categories to use with hue
categories = ['Classification'] * len(models_cls) + ['Regression'] * len(models_reg) + ['Avg'] * 2
plot_df = pd.DataFrame({
    'Model': final_models,
    'Score': final_scores,
    'Category': categories
})

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=plot_df, x='Model', y='Score', hue='Category', dodge=False, palette='Set2', legend=False)

# Add score labels
for i, row in plot_df.iterrows():
    plt.text(i, row['Score'] + 0.02, f"{row['Score']:.2f}", ha='center', fontsize=10)

plt.title("Classification vs Regression Model Comparison")
plt.ylabel("Score")
plt.ylim(0, 1.1)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
